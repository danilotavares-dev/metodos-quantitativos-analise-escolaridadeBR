```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Carregar Dados

```python
df = pd.read_excel('datasets/nivel_escolaridade.xlsx', header=[1, 2], index_col=0, skipfooter=1)

df.columns = pd.MultiIndex.from_tuples([
    (str(col[0]).strip(), str(col[1]).strip()) 
    for col in df.columns
])
```

# Análise de Brasil

```python
df_brasil = df.loc[['Brasil']]

homens = df_brasil['Homens'].iloc[0]
mulheres = df_brasil['Mulheres'].iloc[0]
```

```python
df_compare = pd.DataFrame({'Homens': homens, 'Mulheres' : mulheres})
df_compare
```

```python
df_temp = df_compare.reset_index()

df_final = df_temp.melt(id_vars='index', var_name='Sexo', value_name='Porcentagem')
```

```python
df_final
```

```python
plt.figure(figsize=(20, 10))
fig = sns.barplot(data=df_final, x='index', y='Porcentagem', hue='Sexo')

plt.show()
```

### Conclusão

Após a análise de comparação entre "Mulheres" e "Homens" em nível nacional, descobrimos que o sexo feminino tem um indice de formação superior ao dos homens. Obtendo como resultado (em porcentagem):

Mulheres:
- Sem instrução e fundamental incompleto    ->  30.228135
- Fundamental completo e médio incompleto   ->  14.416825
- Médio completo e superior incompleto      ->  36.288430
- Superior completo                         ->  19.066610

Homens:
- Sem instrução e fundamental incompleto    ->  33.904177
- Fundamental completo e médio incompleto   ->  16.521064
- Médio completo e superior incompleto      ->  35.350601
- Superior completo                         ->  14.224158
