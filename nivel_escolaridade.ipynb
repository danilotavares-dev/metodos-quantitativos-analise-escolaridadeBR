## Análise de Estados

```python
df.index = df.index.astype(str).str.strip()

lista_ufs = [
    'Rondônia', 'Acre', 'Amazonas', 'Roraima', 'Pará', 'Amapá', 'Tocantins',
    'Maranhão', 'Piauí', 'Ceará', 'Rio Grande do Norte', 'Paraíba', 'Pernambuco', 
    'Alagoas', 'Sergipe', 'Bahia', 'Minas Gerais', 'Espírito Santo', 
    'Rio de Janeiro', 'São Paulo', 'Paraná', 'Santa Catarina', 
    'Rio Grande do Sul', 'Mato Grosso do Sul', 'Mato Grosso', 'Goiás', 'Brasília (DF)'
]

df_ufs = df[df.index.isin(lista_ufs)]

df_ufs_superior = df.loc[lista_ufs, ('Total', 'Superior completo')]
```

```python
df_ufs_superior.describe()
```

```python
estatisticas = df_ufs_superior.describe()[['mean', '50%', 'std', 'min', 'max']]
estatisticas['amplitude'] = estatisticas['max'] - estatisticas['min']

estatisticas = estatisticas.rename({
    'mean': 'Média',
    '50%': 'Mediana',
    'std': 'Desvio Padrão',
    'min': 'Mínimo',
    'max': 'Máximo',
    'amplitude': 'Amplitude'
})

df_estatisticas = estatisticas.to_frame(name='Valores (%)')
print('\n')
display(df_estatisticas)
```

```python
estado_maior = df_ufs_superior.idxmax()
valor_maior = df_ufs_superior.max()

estado_menor = df_ufs_superior.idxmin()
valor_menor = df_ufs_superior.min()

print("\n--- DESTAQUES NACIONAIS ---")
print(f"Destaque Positivo (Maior %): {estado_maior} com {valor_maior:.2f}%")
print(f"Destaque Negativo (Menor %): {estado_menor} com {valor_menor:.2f}%\n")

df_ordenado = df_ufs_superior.sort_values(ascending=False).to_frame('Superior Completo (%)')

print("--- RANKING COMPLETO DOS ESTADOS ---")
display(df_ordenado)
```

A análise estatística das Unidades da Federação ratifica a acentuada disparidade educacional no território brasileiro. Os dados revelam uma amplitude de 21,10 pontos percentuais entre os extremos da lista. Enquanto o estado do Maranhão apresenta o menor índice de formação superior (12,06%), Brasília (DF) destaca-se com 33,16%, um valor demasiadamente superior à média nacional.

Esse comportamento do Distrito Federal pode ser atribuído à sua característica de centro administrativo, que atrai uma força de trabalho com alta qualificação para o setor público. No entanto, ao observar o ranking completo, nota-se que a maioria dos estados do Norte e Nordeste ocupa as posições inferiores, enquanto o Sul e Sudeste dominam o topo, confirmando que a escolaridade no Brasil ainda está fortemente atrelada ao desenvolvimento socioeconômico regional.

## Análise de Município

#### Município escolhido: Goiás

```python
df
```

```python
df_goias = df[df.index.str.contains(r'\(GO\)$', regex=True)]

superior_goias = df_goias[('Total', 'Superior completo')]

stats_go = superior_goias.describe()[['mean', '50%', 'std', 'min', 'max']]
stats_go['amplitude'] = stats_go['max'] - stats_go['min']

stats_go = stats_go.rename({
    'mean': 'Média', '50%': 'Mediana', 'std': 'Desvio Padrão',
    'min': 'Mínimo', 'max': 'Máximo', 'amplitude': 'Amplitude'
})

print("\n--- ESTATÍSTICA MUNICIPAL: GOIÁS ---")
display(stats_go.to_frame('Valores (%)'))

plt.figure(figsize=(12, 6))
sns.histplot(superior_goias, bins=15, kde=True, color='seagreen')

plt.title('Distribuição do Ensino Superior nos Municípios de Goiás', fontsize=15)
plt.xlabel('Percentual de Superior Completo (%)', fontsize=12)
plt.ylabel('Frequência (Quantidade de Municípios)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
print("\nTOP 5 MUNICÍPIOS (GO):")
display(superior_goias.sort_values(ascending=False).head(5))

print("\nBOTTOM 5 MUNICÍPIOS (GO):")
display(superior_goias.sort_values(ascending=True).head(5))
```

```python
top_5 = superior_goias.sort_values(ascending=False).head(5).to_frame('Percentual')
top_5['Status'] = 'Destaque Positivo'

bottom_5 = superior_goias.sort_values(ascending=True).head(5).to_frame('Percentual')
bottom_5['Status'] = 'Destaque Negativo'

df_mapa_final = pd.concat([top_5, bottom_5])
df_mapa_final.index = df_mapa_final.index.str.replace(r' \(GO\)$', '', regex=True).str.strip()
df_mapa_final.index.name = 'Cidade'

df_mapa_final['Estado'] = 'Goiás'

print(df_mapa_final.reset_index().to_string(index=False))
```
